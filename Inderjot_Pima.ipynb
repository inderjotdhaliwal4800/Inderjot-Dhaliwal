```python
# creating list for KNN
k = list(range(1,40,2))
# creating list of cv scores
cv = []

# perform 10-fold cross validation
for i in k:
    knn = neighbors.KNeighborsClassifier(n_neighbors=i)
    scores = model_selection.cross_val_score(knn, x_train, y_train, cv=10, scoring='accuracy')
    cv.append(scores.mean())
```

```python
# changing to misclassification error
error = [1 - x for x in cv]

plt.figure()
plt.figure(figsize=(15,10))
plt.title('The optimal number of neighbors', fontsize=20, fontweight='bold')
plt.xlabel('Number of Neighbors K', fontsize=15)
plt.ylabel('Misclassification Error', fontsize=15)
sns.set_style("whitegrid")
plt.plot(k,error)
```

```python
## finding best k value
best_k_value = k[error.index(min(error))]
print("The optimal number of neighbors is %d." % best_k_value)
```

```python
## Creating neW KNN model for best accuracy

knn_new = neighbors.KNeighborsClassifier(n_neighbors=best_k_value)
```

```python
knn_new.fit(x_train, y_train)
```

```python
## Predicting results
predictions_2 = knn_new.predict(x_test)
```

```python
print("Confusion matrix: \n", confusion_matrix(y_test, predictions_2))
```

```python
print("Score {:.2f}".format(knn_new.score(x_test, y_test)))
```

```python
print("Accuracy {:.2f} ".format(accuracy_score(y_test, predictions_2)))
```
